```python
amended_data = amended_data.replace(['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['0','1','1','1','2','2','2'])
```

#PCA Visualization for LSD

```python
from sklearn.decomposition import PCA
```

```python
features = ['Age','Education','Ethnicity','Nscore', 'Escore', 'Oscore', 'Ascore', 'Csore', 'Impulsive','SS']
```

```python
#Separating out the features
x = amended_data.loc[:,features].values
```

```python
#Separating out the target
y = amended_data.loc[:,['LSD']].values
```

```python
pca = PCA(n_components = 2)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, amended_data[['LSD']]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['0','1','2']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['LSD'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
